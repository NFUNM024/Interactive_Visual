```python
from pyecharts.charts import Bar,Tab,Line,Map,Timeline,Grid,Scatter
from pyecharts import options as opts 
import pandas as pd 
import numpy as np

df2 = pd.read_csv("gdp.csv",encoding='utf-8') 

def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2013, 2018):
        map0 = (
            Map()
            .add(
                "", list(zip(list(df2.地区.unique()),list(df2["{}".format(i)]))), "world",is_map_symbol_show = False
            )
            .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
            .set_global_opts(
                title_opts=opts.TitleOpts(title="2013年-2017年世界各地区GDP总量".format(i),subtitle="",
                                         subtitle_textstyle_opts=opts.TextStyleOpts(color="black",font_size=16,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(max_=15000000000000),
               
            
            )
        )
        tl.add(map0, "{}年".format(i))
    return tl
Timeline_map = timeline_map()
Timeline_map.render('世界各地区GDP总量.html')
```

```python
df = pd.read_csv("lvyou.csv",encoding='utf-8') 

def timeline_bar() -> Timeline:
    x = list(df.地区)
    tl = Timeline()
    for i in range(2013, 2018):
        bar = (
            Bar()
            .add_xaxis(x)
            .add_yaxis("世界各地区旅游总收入", list(df["{}".format(i)]))
            .add_yaxis("世界各地区GDP总量", list(df2["{}".format(i)]))
            .reversal_axis()
            .set_global_opts(title_opts=opts.TitleOpts("{}年".format(i)),
                            datazoom_opts=opts.DataZoomOpts(orient="vertical",range_start="50",range_end="60"))
            .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False),
        )
        )
        tl.add(bar, "{}年".format(i))
    return tl

Timeline_bar = timeline_bar()
Timeline_bar.render('世界各地区旅游总收入与GDP总量对比.html')
```

```python
import pandas as pd
df3=pd.read_csv("pm2.5.csv",index_col=0)
df3.head()
```

```python
df3.columns = [int(x) for x in df3.columns]
df3.index = ['Australia','Canada','China','France','United Kingdom','Japan','Russia','United States']
df3.head()
```

```python
import plotly as py
import plotly.graph_objs as go

中国 = go.Scatter(
    x=[pd.to_datetime('01/01/{y}'.format(y=x), format="%m/%d/%Y") for x in df3.columns.values],
    y=df3.loc["China",:].values,
    name = "中国"
)

美国 = go.Scatter(
    x=[pd.to_datetime('01/01/{y}'.format(y=x), format="%m/%d/%Y") for x in df3.columns.values],
    y=df3.loc["United States",:].values,
    name = "美国"
)
澳大利亚 = go.Scatter(
    x=[pd.to_datetime('01/01/{y}'.format(y=x), format="%m/%d/%Y") for x in df3.columns.values],
    y=df3.loc["Australia",:].values,
    name = "澳大利亚"
)
法国 = go.Scatter(
    x=[pd.to_datetime('01/01/{y}'.format(y=x), format="%m/%d/%Y") for x in df3.columns.values],
    y=df3.loc["France",:].values,
    name = "法国"
)
英国 = go.Scatter(
    x=[pd.to_datetime('01/01/{y}'.format(y=x), format="%m/%d/%Y") for x in df3.columns.values],
    y=df3.loc["United Kingdom",:].values,
    name = "英国"
)
日本 = go.Scatter(
    x=[pd.to_datetime('01/01/{y}'.format(y=x), format="%m/%d/%Y") for x in df3.columns.values],
    y=df3.loc["Japan",:].values,
    name = "日本"
)
俄罗斯 = go.Scatter(
    x=[pd.to_datetime('01/01/{y}'.format(y=x), format="%m/%d/%Y") for x in df3.columns.values],
    y=df3.loc["Russia",:].values,
    name = "俄罗斯"
)

layout = dict(xaxis=dict(rangeselector=dict( buttons=list([
                                                dict(count=3,
                                                     label="2年",
                                                     step="year",
                                                     stepmode="backward"),
                                                dict(count=5,
                                                     label="4年",
                                                     step="year",
                                                     stepmode="backward"),
                                                dict(count=10,
                                                     label="6年",
                                                     step="year",
                                                     stepmode="backward"),
                                                dict(count=20,
                                                     label="8年",
                                                     step="year",
                                                     stepmode="backward"),
                                                dict(step="all")
                                            ])),
                         rangeslider=dict(bgcolor="#EEB4B4"),
                         title='年份'
                        ),
              yaxis=dict(title='PM2.5污染量'),
              title="PM2.5污染量对比"               
             )

shouru = dict(data=[中国,美国,澳大利亚,法国,英国,日本,俄罗斯], layout=layout) 

py.offline.plot(shouru,filename = 'PM2.5污染量对比.html')
```
